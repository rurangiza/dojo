## 1st approach
empty: head -> None, tail -> None
full:   head -> firstnode -> none -> lastnode -> None
        tail -> lastnode

```python
class Node:
    def __init__(self, value=-1):
        self.val = value
        self.next = None
        self.prev = None

class Deque:
    
    def __init__(self):
        self.head = Node()
        self.tail = Node()
        self.prevtail = None
        self.size = 0

    def isEmpty(self) -> bool:
        return self.size == 0

    def append(self, value: int) -> None:
        newnode = Node(value)
        if self.isEmpty():
            self.head.next = newnode
            self.tail.next = newnode
        else:
            lastnode = self.tail.next
            lastnode.next = newnode
            newnode.prev = lastnode
            self.tail.next = newnode
        self.size += 1

    def appendleft(self, value: int) -> None:
        newnode = Node(value)
        if self.isEmpty():
            self.head.next = newnode
            self.tail.next = newnode
        else:
            firstnode = self.head.next
            newnode.next = firstnode
            firstnode.prev = newnode
            self.head.next = newnode
        self.size += 1

    def pop(self) -> int:
        if self.isEmpty():
            return -1
        val = self.tail.next.val
        if self.size == 1:
            self.head.next = None
            self.tail.next = None
        else:
            oldlast = self.tail.next
            newlast = oldlast.prev
            newlast.next = None
            self.tail.next = newlast
        self.size -= 1
        return val

    def popleft(self) -> int:
        if self.isEmpty():
            return -1
        val = self.head.next.val
        if self.size == 1:
            self.head.next = None
            self.tail.next = None
        else:
            oldfirst = self.head.next
            newfirst = oldfirst.next
            self.head.next = newfirst
            newfirst.prev = None
        self.size -= 1
        return val
```

```python
class Node:
    def __init__(self, value):
        self.val = value
        self.next = None
        self.prev = None

class Deque:
    
    def __init__(self):
        self.head = Node(-1)
        self.tail = Node(-1)
        self.head.next = self.tail
        self.tail.prev = self.head

    def isEmpty(self) -> bool:
        return self.head.next == self.tail

    def append(self, value: int) -> None:
        newnode = Node(value)
        lastnode = self.tail.prev

        lastnode.next = newnode
        newnode.prev = lastnode
        newnode.next = self.tail
        self.tail.prev = newnode
        

    def appendleft(self, value: int) -> None:
        newnode = Node(value)
        firstnode = self.head.next

        firstnode.prev = newnode
        newnode.next = firstnode
        newnode.prev = self.head
        self.head.next = newnode

    def pop(self) -> int:
        if self.isEmpty():
            return -1
        oldlast = self.tail.prev
        newlast = oldlast.prev

        newlast.next = self.tail
        self.tail.prev = newlast

        return oldlast.val

    def popleft(self) -> int:
        if self.isEmpty():
            return -1
        oldfirst = self.head.next
        newfirst = oldfirst.next
        newfirst.prev = self.head
        self.head.next = newfirst
        return oldfirst.val
```
